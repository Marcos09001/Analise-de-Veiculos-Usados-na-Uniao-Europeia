# Introdução ao problema

O mercado europeu de veículos usados movimenta bilhões de euros anualmente e apresenta forte competitividade, com variações significativas de preço influenciadas por múltiplos fatores técnicos e estruturais.

Neste projeto, realizo uma Análise Exploratória de Dados (EDA) em um conjunto de dados de veículos usados da União Europeia com o objetivo de identificar:

1. Principais fatores que impactam o preço;

2. Padrões de depreciação ao longo do tempo;

3. Diferenças de valor entre tipos de combustível e categorias;

4. Segmentações estratégicas de mercado.

O projeto envolve etapas completas de:

1. Limpeza e tratamento de dados;

2. Tratamento de valores ausentes;

3. Padronização de tipos;

4. Criação de novas variáveis (feature engineering);

5. Análise estatística descritiva;

6. Geração de insights orientados a negócio.

A proposta vai além da visualização gráfica, buscando transformar dados brutos em insights estratégicos aplicáveis a decisões comerciais, precificação e modelagem preditiva.

Este estudo demonstra a aplicação prática de técnicas de análise de dados voltadas para problemas reais de mercado, reforçando competências em manipulação de dados, raciocínio estatístico e geração de valor a partir de informação estruturada.

```python
# Tratamento dos valores
df.fillna({'Model':'Not informed'}, inplace=True)
df.fillna({'Body': 'Not informed'}, inplace=True)
df.fillna({'Country': 'Not informed'}, inplace=True)
df.fillna({'Fuel_Type': 'Not informed'}, inplace=True)
df.fillna({'Drivetrain': 'Not informed'}, inplace=True)
df.fillna({'Gearbox': 'Not informed'}, inplace=True)
df.fillna({'Gears': df['Gears'].median()}, inplace=True)
df.fillna({'Engine_Size_cc': df['Engine_Size_cc'].median()}, inplace=True)
df.fillna({'Cylinders': df['Cylinders'].median()}, inplace=True)
df.fillna({'Seats': df['Seats'].median()}, inplace=True)
df.fillna({'Doors': df['Doors'].median()}, inplace=True)
df.fillna({'Color': 'Not informed'}, inplace=True)
df.fillna({'Upholstery': 'Not informed'}, inplace=True)
df.fillna({'Previous_Owners': df['Previous_Owners'].median()}, inplace=True)
df.fillna({'Seller': 'Not informed'}, inplace=True)
df.fillna({'Power_hp': df['Power_hp'].mean()}, inplace=True)
df.fillna({'Full_Service_History': 'Not informed'}, inplace=True)
df.fillna({'Non_Smoker_Vehicle': 'Not informed'}, inplace=True)
```

# Análise Exploratória de Dados (EDA)

```python
# Adição de uma nova coluna identificando a categoria dos carros pegando como base o preço
def categorizar(Preco):
  if Preco <= 10000:
    return 'Compact'
  elif Preco <= 40000:
    return 'Average/Hatchback'
  elif Preco <= 45000:
    return 'SUV'
  else:
    return 'Premium'

df['Category'] = df['Price'].apply(categorizar)
```

```python
# Verificar a alteração no dataframe
df
```
